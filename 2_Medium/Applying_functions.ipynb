```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = pd.DataFrame(dataset['train'])
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']).dt.strftime('%B')
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
type(df1['job_skills'][1])
```

```python
import ast
```

```python
def cl_skills(x):
    if pd.notna(x):
        return ast.literal_eval(x)

df1['job_skills'] = df1['job_skills'].apply(cl_skills)
```

```python
df1.head()
```

```python
def interest_sal(row):
    if row['job_title_short'] == 'Data_Analyst':
        return row['salary_year_avg'] * 1.1
    else:
        return row['salary_year_avg'] * 1.5

df1['Interest_salary'] = df1.apply(interest_sal, axis=1)
```
