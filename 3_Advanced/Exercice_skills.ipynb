```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.tail()
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
df['job_posted_year'] = df['job_posted_date'].dt.year
```

```python
df.tail()
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded.head()
```

```python
import numpy as np
job_name = "Data Analyst"
top_5 = 5


job_filtered = df_exploded[df_exploded['job_title_short'] == job_name]
job_filtered['month_n'] = job_filtered['job_posted_date'].dt.month
```

```python
job_filtered.head
```

```python
job_ms = job_filtered.pivot_table(index='month_n', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
job_ms.loc['total'] = job_ms.sum()
```

```python
job_ms_t = job_ms.loc['total'].sort_values(ascending=False).head(top_5).index.tolist()
job_ms_top = job_ms[job_ms_t]

job_ms_top.drop('total', inplace=True)
```

```python
job_ms_top
```

```python
job_ms_top.plot(kind = 'line')
```

```python
job_ms_top = job_ms_top.reset_index()
```

```python
job_ms_top['month_n'] = pd.to_datetime(job_ms_top['month_n'], format='%m')
```

```python
job_ms_top['month_n'] = job_ms_top['month_n'].dt.strftime('%B')
```

```python
job_ms_top
```

```python
job_ms_top.set_index('month_n', inplace=True)
```

```python
job_ms_top
```

```python
job_ms_top.plot(kind = 'line')
plt.xticks(rotation=45)
plt.title('top_5')
plt.legend()
```
