```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = pd.DataFrame(dataset['train'])
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US.head()
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist', bins=30, color='blue', edgecolor='black', figsize=(10,6))
ax = plt.gca()
ax.xaxis.set_major_formatter(lambda x, pos: f'{int(x/1000)}k')
plt.title('Distribution of Average Yearly Salary for Data Analysts in the US')
plt.xlabel('Average Yearly Salary (USD)')
plt.ylabel('Number of Job Postings')
plt.xlim(0, 250000)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
