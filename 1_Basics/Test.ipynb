```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = pd.DataFrame(dataset['train'])
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_year'] = df['job_posted_date'].dt.year
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
df_altered = df.copy()
df_altered.head()
```

```python
median_job_hsalary = df_altered['salary_hour_avg'].median()
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_job_hsalary)

df_altered['salary_hour_avg']
```

```python
job_country_hsalary = df_altered.pivot_table(values = 'salary_hour_avg', index = 'job_country', columns = 'job_title_short', aggfunc = 'median')
```

```python
top_countries_1 = df_altered['job_country'].value_counts().nlargest(10).index

top_countries_2 = df_altered['job_country'].value_counts().head(5).index
```

```python
df_job_country_hsalary=job_country_hsalary.loc[top_countries_2]
```

```python
job_titles =['Business Analyst','Data Scientist','Data Engineer']
df_job_country_hsalary=df_job_country_hsalary[job_titles]
df_job_country_hsalary
```
